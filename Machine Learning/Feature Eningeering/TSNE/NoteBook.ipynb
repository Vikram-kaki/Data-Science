# TSNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file1 = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Bank%20Note/bank_note_data.csv')
```

```python
file1.head()
```

```python
vars = file1.columns[:-1].to_list()
target = file1.columns[-1]
```

```python
vars
```

```python
target
```

```python
file1[vars]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaledData = scaler.fit_transform(file1.loc[:, vars])
```

```python
scaledData
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=50)

xtbse = tsne.fit_transform(scaledData)
```

```python
xtbse
```

```python
target
```

```python
sns.scatterplot(x = xtbse[:, 0], y = xtbse[:, 1], c = file1[target])
plt.show()
```
