# **K-Means Clustering**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs


X, y= make_blobs(5000, 2, centers=20) # type: ignore
```

```python
from sklearn.cluster import KMeans

k_means = KMeans(n_clusters=20)
```

```python
k_means.fit(X, y)
```

```python
centers = k_means.cluster_centers_
```

```python
labels = k_means.labels_
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c= labels)
plt.scatter(x = centers[:, 0], y = centers[:, 1], c = 'orange', s = 500, marker = '+')
plt.show()
```
