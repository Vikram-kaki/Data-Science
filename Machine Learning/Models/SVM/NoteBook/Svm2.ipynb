```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file1 = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Insurance/insurance.csv')
```

```python
file1.head()
```

```python
file1['male'] = pd.get_dummies(file1['sex'], drop_first=True).astype('int')
```

```python
file1
```

```python
file1['smoker'] = pd.get_dummies(file1['smoker'], drop_first = True).astype('int')
```

```python
file1
```

```python
Region = pd.get_dummies(file1['region']).astype('int')
```

```python
Region
```

```python
file1 = pd.concat([file1, Region], axis = 1)
```

```python
file1
```

```python
import statsmodels.api as sm
```

```python
X = file1.drop(['sex', 'region', 'charges', 'northeast', 'northwest', 'southeast', 'southwest'], axis = 1)
```

```python
X
```

```python
y = file1['charges']
```

```python
X_constant = sm.add_constant(X)
```

```python
model = sm.OLS(y, X_constant)
```

```python
result = model.fit()
```

```python
print(result.summary())
```

```python
X.drop('male', inplace = True, axis = 1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X['age'] = scaler.fit_transform(X['age'].values.reshape(-1, 1))
X['bmi'] = scaler.fit_transform(X['bmi'].values.reshape(-1, 1))
```

```python
X
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.svm import SVR


regressor = SVR(kernel = 'rbf', C = 10000, gamma = 0.1)
```

```python
y_train
```

```python
regressor.fit(X_train, y_train)
```

```python
predict = regressor.predict(X_test)
```

```python
predict
```

```python
y = y_test.iloc[:].values
```

```python
y
```

```python
for i in range(len(predict)):
    print(f" {y[i]} :  {predict[i]}")
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, predict)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100, 1000, 10000, 100000],
    'gamma': [0.001,0.01, 0.1, 1, 10]
}

# Create SVR model
svr = SVR(kernel='rbf')

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=svr, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Get the best model
best_svr = grid_search.best_estimator_

# Predictions
y_pred_train_best = best_svr.predict(X_train)
y_pred_test_best = best_svr.predict(X_test)

# Evaluating the model
# train_rmse_best = np.sqrt(mean_squared_error(y_train, y_pred_train_best))
# test_rmse_best = np.sqrt(mean_squared_error(y_test, y_pred_test_best))
# print("Train RMSE (Best Model):", train_rmse_best)
# print("Test RMSE (Best Model):", test_rmse_best)
```
