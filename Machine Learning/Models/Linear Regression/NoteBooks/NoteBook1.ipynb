```python
def forwardPropagation(train_input, m, c):
    predictions = np.multiply(train_input , m) + c
    return predictions


def costFunction(train_output, predictions):
    cost = np.mean((train_output - predictions) ** 2) * 0.5
    return cost
```

```python
def backwardPropagation(train_input, train_output,  predictions):
    derivatives = dict()
    
    df = predictions - train_output
    dm = np.mean(np.multiply(df, train_input))
    dc = np.mean(df)
    
    derivatives['dm'] = dm
    derivatives['dc'] = dc
    
    return derivatives
```

```python
def updateParameters(parameters, derivatives, learning_rate):
    parameters['m'] = parameters['m'] - (learning_rate*derivatives['dm'])
    parameters['c'] = parameters['c'] - (learning_rate*derivatives['dc'])
    
    return parameters
```

```python
def train(train_input, train_output, learning_rate, iters):
    
    parameters = dict()
    loss = []
    
    parameters['m'] = np.random.uniform(0, 1)
    parameters['c'] = np.random.uniform(0, 1)
    
    for i in range(iters):
        predictions = forwardPropagation(train_input=train_input, m = parameters['m'], c = parameters['c'])
        
        cost = costFunction(train_output=train_output, predictions=predictions)
        loss.append(cost)
        
        derivatives = backwardPropagation(train_input, train_output, predictions)
        
        parameters = updateParameters(parameters, derivatives, learning_rate)
        
        
    return parameters, loss
```

```python
parameters, loss = train(training_input, training_output, 0.0001, 100)
```
