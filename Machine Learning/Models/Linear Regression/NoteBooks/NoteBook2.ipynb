```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
file1 = pd.read_csv('https://raw.githubusercontent.com/AshishJangra27/Machine-Learning-with-Python-GFG/main/Linear%20Regression/data_for_lr.csv')
```

```python
file1.dropna(inplace = True)
```

```python
file1
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(file1.iloc[:,:-1].values, file1.iloc[:,1:].values, test_size=0.3)
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
X_train
```

```python
lr_model.predict(X_train)
```

```python
predictions = lr_model.predict(X_test)
for i in range(len(X_test)):
    print(f"{X_test[i]} : {y_test[i]}, {predictions[i]}")
```

```python
from sklearn.metrics import mean_squared_error


mean_squared_error(y_test, predictions)
```
