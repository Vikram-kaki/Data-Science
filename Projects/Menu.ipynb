```python
# Let's get the numerical data.

numerical_data = file1.select_dtypes(include=['number'])
numerical_data.columns
```

### Let's find the mean value of numerical_columns associated with each category

```python
groupByData = file1.groupby('Category')
# let's find the mean of each of the numerical value found in each category.

for i in numerical_data.columns:
    print("*"*100)
    mean_values = groupByData[i].mean()
    print("-"*100)
    max_value = groupByData[i].mean().max()
    min_value = groupByData[i].mean().min()
    max_index = groupByData[i].mean().idxmax()
    min_index = groupByData[i].mean().idxmin()
    print(f"{i} :\n{mean_values}")
    print(f"Max Mean {i} : {max_value} ({max_index})")
    print(f"Min Mean {i} : {min_value} ({min_index})")
    print("-"*100)
```
