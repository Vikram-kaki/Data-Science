#### Loops in Python

```python
# for loop
for i in range(5):
    print(i)
```

```python
# while loop
i = 0
while i < 5:
    print(i)
    i += 1
```

```python
# Break and Continue in loops.
for i in range(10):
    if i == 5:
        break
    print(i)

print()
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
# A program to print the table of a given number by taking the input from the user and the second the input as the number of multiples.

num = int(input("Enter the number: "))
mul = int(input("Enter the number of multiples: "))
for i in range(1, mul+1):
    print(num, "x", i, "=", num*i)
```

```python
# range() function in python.

# range(start, stop, step)
# start - starting number of the sequence
# stop - generate numbers up to, but not including this number
# step - difference between each number in the sequence

values = list(range(10, 20, 2))
print(values)
```

```python
# A program to find the smallest divisor of a number such that the divisor is greater than 1


def smallestDivisor(n):
    for i in range(2, n+1):
        if n % i == 0:
            return i

print(smallestDivisor(15))
```

```python
# A program to print the first N numbers tables by taking the n as input from the user.


def print_tables(n):
    for i in range(1, n+1):
        for j in range(1, 11):
            print(f"{i} X {j} == {i*j}")
        print()


n = int(input("Enter the N value : "))
print_tables(n)
```

#### List Comprehension in Python.

```python
var1 = [x for x in range(11)]
print(var1)


var2 = [x**2 for x in range(11)]
print(var2)


var3 = [x for x in range(11) if x%2==0]
print(var3)


var4 = [[x for x in range(11)] for x in range(5)]
print(var4)


print([num for row in var4 for num in row])
```

#### Dict Comprehension in Python

```python
dic1 = {x : chr(x) for x in range(32, 1000)}
print(dic1)


dic2 = {x : x**2 for x in range(1,101)}
for i in dic2.keys():
    print(dic2.get(i))
    
    
    
dic3 = {i: {x : x**i for x in range(1,101)} for i in range(1,5)}
for i in dic3.keys():
    print(i, dic3.get(i), end='\n\n*******************************************************\n')
```
