### Sine and Cosine Waves functions graphs.

##### Sin Graph

```python
s1 = np.arange(1,360*3)
sin_x = np.sin(s1 * (np.pi/180))

plt.plot(s1, sin_x, color='red', linestyle = 'dotted', linewidth = 1.3)
plt.xlabel('Angle')
plt.ylabel('Sin value')
plt.title("Sin Graph")
plt.xticks(range(0,(360*3)+1,90))
plt.show()
```

##### Cos Graph

```python
s2 = np.arange(1,360*3)

cos_x = np.cos(s2*(np.pi/180))


plt.plot(s2,cos_x, color = 'black')
plt.xlabel('Angle')
plt.ylabel('Cos Value')
plt.title('Cos Graph')
plt.xticks(range(0,721,90))
plt.show()
```

##### Drawing both the graphs in a single plot.

```python
plt.figure(figsize=(10,5)) # This increases or decreases the plot size
fig = plt.figure()
ax = plt.axes()
ax.set_facecolor('white')
plt.plot(s1, sin_x, color = 'skyblue', linestyle = 'dotted', linewidth = 1.5, label = 'Sin Wave')
plt.plot(s2, cos_x, color = 'red', linestyle = 'dotted', linewidth = 1.5, label = 'Cos Wave')
plt.xticks(range(0, (360*3), 90))
plt.legend()
plt.grid(color='black')
plt.xlabel('Angle')
plt.ylabel('Value')
plt.title('Sin Vs Cos')
fig.savefig('Sin vs Cos')
plt.show()
```

## Different Types of plots.

#### Line plot
Line plot is used to plot the data points in a line.


```python
# syntax = plt.plot()

# A single Variable only
p1 = np.linspace(1,100,25)
plt.plot(p1, color='red', linestyle = 'dotted', linewidth = 1.3, label = 'One')
plt.title("One")
plt.show() # Only variable is used, and another axis is just the number of the variable only. No corresponding relationship is exist.

# Double Variable
p2 = np.arange(1,101,4)
plt.plot(p1, p2, color = 'yellow', linestyle = '-', linewidth = 0.9, label = 'Two')
plt.title("Two")
plt.show() # This graph is the relation between two variables.
```

#### Multiple Line Plots.

In a same a plot multiple line plots are shown

```python
s3 = np.random.randint(1,100,20)
v1 = np.linspace(1,100,20)
s4 = np.random.randint(1,100,20)
v2 = np.linspace(1,100,20)
plt.figure(figsize=(20,10))
ax = plt.axes()
ax.set_facecolor('white')
plt.plot(v1, s3, label = 'One')
plt.plot(v2, s4, label = 'Two')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.title('One & Two')
plt.legend()
plt.show()
```

#### Real World Data Sets.

### Adjusting the plot axes.

```python
s5 = np.arange(1,361)
sin_x2 = np.sin(s5*(np.pi/180))

plt.figure(figsize=(60,10))
fig = plt.figure()
ax = plt.axes()
# ax.set_facecolor('White')
plt.plot(s5, sin_x2, linestyle = 'dotted', linewidth = 1.2, label = 'Sin Wave')
plt.title('Sin Wave')
plt.xlabel('Angle')
plt.ylabel('Value')
plt.grid(color = 'black')
plt.legend()
plt.xticks(range(0,361,30))
plt.xlim([0,450])
plt.ylim([-1,1])
# plt.axis('tight')
fig.savefig('Sine Wave')
plt.show()
```

## Scatter Plot
Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. The scatter() method in the matplotlib library is used to draw a scatter plot. Scatter plots are widely used to represent relation among variables and how change in one affects the other.

```python
s6 = np.random.randn(100)
s7 = np.random.randn(100)

fig = plt.figure()
# plt.style.use('seaborn-v0_8-darkgrid')
plt.scatter(s6, s7, marker='.', linewidths=0.2, label = 'R1', color = 'red', s=50, alpha=0.8)
plt.grid(color = 'white')
plt.legend()
fig.savefig('R1')
plt.show()
plt.close()
```

## Histograms And Binnings

A histogram is a graph showing frequency distributions.
It is a graph showing the number of observations within each given interval.

```python
hx1 = np.random.randn(1000000)


fig = plt.figure(figsize = (10, 4))

# ax = plt.axes()
# ax.set_facecolor('white')


plt.hist(hx1, bins=100, label='RF Histo1', edgecolor = 'white', color = 'red', alpha = .86, histtype='barstacked')
# Types of histo-gram are 'step', 'stepfilled', 'barstacked'
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('Random Frequency Histogram')
plt.grid()
plt.legend()
plt.style.use('seaborn-v0_8-darkgrid')
# plt.colorbar()
fig.savefig("Random Histo")
plt.show()
```

#### Multiple Histogram

```python
hx2 = np.random.randn(1000)
hx3 = np.random.randn(1000) + 1
hx4 = np.random.randn(1000) + 2

plt.hist(hx2, bins = 100, alpha = .8,)
plt.hist(hx3, bins = 100, alpha = .8,)
plt.hist(hx4, bins = 100, alpha = .8)
plt.show()
```

### Pie-Charts

```python
laptops = ['Asus', 'Acer', 'Hp', 'Lenovo', 'Mi', 'Realme', 'Mac', 'Samsung']
sales = [19393, 9389, 57683, 12340, 2349, 8899, 34570, 1984]
explode = [0,0,0,0,.8,.9,0,.8]

plt.pie(sales, labels=laptops, shadow=True, autopct='%1.2f%%', explode=explode,
        wedgeprops={'edgecolor': 'white', 'linewidth' : .8, 'antialiased': True})

plt.show()
```

### Sub - Plots

```python
sp1 = np.linspace(0,10,100)


fig, ax = plt.subplots(3,2)
plt.subplots_adjust(hspace=.4, wspace=.4)
y = np.sin(sp1)
ax[0, 0].plot(sp1, y)
ax[0, 0].set_title('Sin Wave')

y = np.cos(sp1)
ax[0, 1].plot(sp1, y)
ax[0, 1].set_title('Cos Wave')

y = np.tan(sp1)
ax[1, 0].plot(sp1, y)
ax[1, 0].set_title('Tan Wave')

y = np.random.randn(100)
ax[1, 1].scatter(sp1, y, marker = '+', s = 2)
ax[1, 1].set_title('Scatter Plot')

y = np.random.rand(100)
ax[2, 0].hist(y, bins = 100)
ax[2, 0].set_title('Histo Graph')

names = ['N1', 'N2', 'N3', 'N4']
km = [4, 3.9, 2.987, 4.231]
ax[2, 1].pie(x = km, labels = names, autopct='%1.1f%%', shadow=True)


plt.show()
```
