# Series

### Creating a pandas a series.

```python
# Using a list.
lst = [1,2,3,4]
# , [5,6,7,8]
s1 = pd.Series(lst, index=[1,2,3,4])
print(s1)
```

```python
# Using a numpy array.
import numpy as np

array1 = np.linspace(1,100,25)
s2 = pd.Series(array1,index=range(1,26))
print(s2)

print("",sep='\n')
array2 = np.random.randint(1,100,20)
s3 = pd.Series(array2, index = range(1,21))
print(s3)
```

```python
# Series Through Dictionary Values.

dic = {'Name':'Vikram','Age':21,'Sex':'Male'}
s4 = pd.Series(dic)
print(s4)
```

### The repeat Function.

```python
s5 = pd.Series([1,2,3,4]).repeat(5)
print(s5)



# Making the indexes to work as a proper way.
s6 = s5.reset_index(drop=True)
print(s6)


s7 = pd.Series([1,2])
print(s7.repeat([2,3]).reset_index(drop=True))
```

### Slicing of dataFrame using iloc and loc

# loc function.

The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc().

```python
print(df3)

print("-"*40)

print(df3.loc[1:3]) # Selecting the rows
print("-"*40)
print(df3.loc[1:3, 'Names':'Age']) # Selecting both the rows and columns
print("-"*40)
# print(df3.loc[1,'Names'])

# Boolean Value.
print(df3.loc[df3.Sex == 'M'])
print(df3.loc[(df3.Sex == 'M') & (df3.Names == 'Vikram')])
print(df3.loc[df3.Age < 45, ['Names','Sex']])
# Updating Values

df3.loc[df3.Names == 'Vikram', 'Age'] = 20
```

### iloc function

The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc()

```python
print(df3)

print("-"*40)

print(df3.iloc[1:3]) # Accessing the rows
print(df3.iloc[1:3, 1:3]) # Accessing the rows and columns.
print(df3.iloc[[1,2]])

print(df3.iloc[0:3, [0,2]]) # Accessing multiple rows and columns data which aren't in proper order.
```

### Mathematical Functions.

```python

np.random.seed(1)


df5 = pd.DataFrame(
    {
        'One':np.random.randint(1,100,10),
        'Two':np.random.randint(1,100,10),
        'Three':np.random.randint(1,100,10),
        'Four':np.random.randint(1,100,10),
        'Five':np.random.randint(1,100,10)
    }, index=range(1,11)
)
print(df5)



# Sum
print(df5.sum()) # Gives the sum column wise sum.
print(df5.sum(axis=1)) # Gives the row wise sum.
print(df5['One'].sum()) # Single Column sum
print(df5.iloc[1,].sum()) # Single row sum

# Mean
print(df5.mean()) # Display the mean of columns.
print(df5.mean(axis=1)) # Display the mean of rows.
print(df5.One.mean()) # Gives the mean of single row.
print(df5.loc[1].mean()) # Gives the mean of single column.

# Median
print(df5.median()) # Display the mean of columns.
print(df5.median(axis=1)) # Display the mean of rows.
print(df5.One.median()) # Gives the mean of single row.
print(df5.loc[1].median()) # Gives the mean of single column.

# Mode
print(df5.mode()) # Display the mode of columns.
print(df5.mode(axis=1)) # Display the mode of rows.
print(df5.One.mode()) # Gives the mode of single row.
print(df5.loc[1].mode()) # Gives the mode of single column.

# Var
print(df5.var()) # Display the var of columns.
print(df5.var(axis=1)) # Display the var of rows.
print(df5.One.var()) # Gives the var of single row.
print(df5.loc[1].var()) # Gives the var of single column.

# Std
print(df5.std()) # Display the std of columns.
print(df5.std(axis=1)) # Display the std of rows.
print(df5.One.std()) # Gives the std of single row.
print(df5.loc[1].std()) # Gives the std of single column.

# Min
print(df5.min()) # Display the min of columns.
print(df5.min(axis=1)) # Display the min of rows.
print(df5.One.min()) # Gives the min of single row.
print(df5.loc[1].min()) # Gives the min of a single column.

# Max
print(df5.max()) # Display the max of columns.
print(df5.max(axis=1)) # Display the max of rows.
print(df5.One.max()) # Gives the max of a single row.
print(df5.loc[1].max()) # Gives the max of a single column.
```

### Pipe Function.

```python
def fun1(i,j):
    return i+j

def fun2(i,j):
    return i-j


def get_mean(col):
    return col.mean()

def get_var(col):
    return col.var()


print(df5)
print(df5.pipe(fun1, 10),end="\n\n")
print(df5.pipe(fun2, 1),end="\n\n")
print(df5.pipe(get_mean),end="\n\n")
print(df5.pipe(get_mean).pipe(get_var))
```

### GroupBy Function

```python
df8 = pd.DataFrame(
    {
        'Name':['Vikram','Ganesh','Shannu','Teja','Harsha','Bharghav'],
        'Sex':['M','M','M','M','F','F'],
        'Age':[21,20,20,21,20,21],
        'City':['Pkl','Pkl','Bmvr','Bmvr','Bmvr','Bmvr'],
        'Job':[4,8,8,3,2,18]
    },index=range(1,7)
)

print(df8)
print('*'*50)


print(df8.groupby('City').count())
print('*'*50)
print(df8.groupby('Job').max())
print('*'*50)
print(df8.groupby('Sex').count())
print('*'*50)
print(df8.groupby(['Sex','Job']).count())
print('*'*50)

print(df8.groupby(['City','Sex']).agg(['count','max','min']))
print('*'*50)
print(df8.groupby('City')['Job'].max())
print('*'*50)
value = df8.groupby('Sex')['Job'].idxmin()
print('*'*50)
print(value)

print(df8.loc[value, 'Name'])


print('*'*50)
print(df8.groupby('Sex')['Job'].mean())
```
