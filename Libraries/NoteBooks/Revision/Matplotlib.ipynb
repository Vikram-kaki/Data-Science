```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
angle = np.arange(0, 721)
sin_value = np.sin(angle * (np.pi/180))
cos_value = np.cos(angle * (np.pi/180))
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(angle, sin_value, linewidth = 0.5, linestyle = '-', color = 'green', label = 'Sin Wave', )
plt.plot(angle, cos_value, linewidth = 0.2, linestyle = 'solid', color = 'red', label = 'Cos Wave', )
plt.xlabel('Angle')
plt.ylabel('Sin Value')
plt.grid()
plt.legend()
plt.xticks(np.arange(0,721,90))
plt.yticks(np.linspace(-1, 1, 9))
plt.title("Sin & Cos Graph")
fig.savefig("Sin & Cos Wave")
plt.show()
```
